````python
import hashlib
import time

class Block:
    def __init__(self, index, previous_hash, timestamp, data, current_hash, miner):
        self.index = index
        self.previous_hash = previous_hash
        self.timestamp = timestamp
        self.data = data
        self.current_hash = current_hash
        self.miner = miner

    def calculate_hash(self):
        block_string = str(self.index) + str(self.previous_hash) + str(self.timestamp) + str(self.data) + str(self.miner)
        return hashlib.sha256(block_string.encode()).hexdigest()

def create_genesis_block():
    return Block(0, "0", time.time(), "Genesis Block", calculate_hash(0, "0", time.time(), "Genesis Block", "Miner_1"), "Miner_1")

def calculate_hash(index, previous_hash, timestamp, data, miner):
    block_string = str(index) + str(previous_hash) + str(timestamp) + str(data) + str(miner)
    return hashlib.sha256(block_string.encode()).hexdigest()

def mine_block(index, previous_hash, data, miner):
    timestamp = time.time()
    current_hash = calculate_hash(index, previous_hash, timestamp, data, miner)
    return Block(index, previous_hash, timestamp, data, current_hash, miner)

def is_valid_block(block, previous_block):
    if previous_block.index + 1 != block.index:
        return False
    if previous_block.current_hash != block.previous_hash:
        return False
    if block.calculate_hash() != block.current_hash:
        return False
    return True

def display_blockchain(chain):
    for block in chain:
        print(f"Block #{block.index} | Miner: {block.miner} | Timestamp: {block.timestamp} | Data: {block.data} | Hash: {block.current_hash}")

def verify_blockchain(chain):
    previous_block = chain[0]
    for block in chain[1:]:
        if not is_valid_block(block, previous_block):
            return False
        previous_block = block
    return True

def create_block(chain, miner):
    new_block = mine_block(len(chain), chain[-1].current_hash, f"Block #{len(chain)} Data", miner)
    chain.append(new_block)
    print(f"Block #{new_block.index} created successfully!")

def verify_single_block(chain, index):
    if index < 0 or index >= len(chain):
        print("Invalid block index!")
        return

    block_to_verify = chain[index]
    previous_block = chain[index - 1] if index > 0 else create_genesis_block()

    if is_valid_block(block_to_verify, previous_block):
        print(f"Block #{index} is valid.")
    else:
        print(f"Block #{index} is not valid.")

def run_test():
    blockchain = [create_genesis_block()]

    while True:
        print("\nOptions:")
        print("1. Create a Block")
        print("2. Verify a Block")
        print("3. Display Whole Chain")
        print("4. Verify Entire Blockchain (with tampering)")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            miner = input("Enter miner name: ")
            create_block(blockchain, miner)
        elif choice == '2':
            index = int(input("Enter block index to verify: "))
            verify_single_block(blockchain, index)
        elif choice == '3':
            print("\nCurrent Blockchain:")
            display_blockchain(blockchain)
        elif choice == '4':
            # Tamper with Block #3 for testing
            blockchain[2].data = "Tampered Data"
            print("\nBlockchain after tampering:")
            display_blockchain(blockchain)
            print("\nVerification Result:", verify_blockchain(blockchain))
            print()
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    run_test()
````
